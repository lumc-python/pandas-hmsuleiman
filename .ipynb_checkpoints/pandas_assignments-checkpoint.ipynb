### Create and display a one-dimensional array-like object containing an array of data. 

```python
import pandas as pd 
import numpy as np
```

```python
# Creating a series 
s = pd.Series(range(1,11))
s
```

### Convert a Panda module Series to Python list and it's type.

```python
l = list(s) # convert panda series 
print(l) # print list
print(type(l)) # print list type
```

### Add, subtract, multiple and divide two Pandas Series.
Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 9]

```python
s1 = pd.Series(range(1,11))
s2 = pd.Series(range(12,22))
print(s1, s2)

#add
print(s1 + s2)

#subtract
print(s2 - s1)

#multiple
print(s1 * s2)
```
